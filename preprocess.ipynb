```python
df_1.columns = df_1.columns.str.strip() #remove spaces

#all bad rows from SCATS Number column
bad_rows = df_1[df_1['SCATS Number'].isin(bad_sites)].index

df_1_edit=df_1.drop(bad_rows)
print ("rows of bad sites dropped")
```

```python
time_stamp_col = [f'V{str(i).zfill(2)}' for i in range(96)]
time_stamp_col
```

```python
#rearranging the data frame
df_melted = df_1_edit.melt(
    id_vars=['SCATS Number', 'Location', 'Date'],
    value_vars=time_stamp_col,
    var_name='Interval',
    value_name='Volume'
)
```

```python
df_melted.head(20)
```

```python
df_melted['Date'] = pd.to_datetime(df_melted['Date']).dt.date
df_melted
```

```python
#v00 to v95 need to be converted to t
interval_to_time = {
    f'V{str(i).zfill(2)}': f'{(i * 15) // 60:02d}:{(i * 15) % 60:02d}'
    for i in range(96)
}
df_melted['Time'] = df_melted['Interval'].map(interval_to_time)
df_melted.head(20)
```

```python
df_melted.tail(20)
```

```python
df_melted['DateTime'] = pd.to_datetime(df_melted['Date'].astype(str) + ' ' + df_melted['Time'])
df_melted.head(20)
```
